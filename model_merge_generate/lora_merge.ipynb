```python
import transformers 
import peft 
from peft import PeftConfig, PeftModelForCausalLM
from transformers import AutoModelForCausalLM
```

```python
peft_model_id = "/largespace/tydata/code_optimization/cpp/saved_models/pie-gem5-by-user-cpp_deepseekcoder-7b_sft_0418/"

config = PeftConfig.from_pretrained(peft_model_id)
print(f"config = {config}")
```

```python
model = AutoModelForCausalLM.from_pretrained(config.base_model_name_or_path, torch_dtype="auto")
for name, param in model.named_parameters():
    print(
                "name: {}, dtype: {}, device: {}, trainable: {}".format(
                    name, param.dtype, param.device, param.requires_grad
                )
            )
```

```python
model = PeftModelForCausalLM.from_pretrained(model, peft_model_id)
model = model.merge_and_unload()
out_dir = "/largespace/tydata/code_optimization/cpp/saved_models/pie-gem5-by-user-cpp_deepseekcoder-7b_sft_0418/full_model/"
model.save_pretrained(out_dir)
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(config.base_model_name_or_path)
tokenizer.save_pretrained(out_dir)
```
